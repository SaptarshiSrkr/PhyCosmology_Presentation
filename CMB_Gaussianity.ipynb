```python
import matplotlib.pyplot as plt
import numpy as np
import healpy as hp
import camb

from matplotlib import cm
from scipy.stats import norm

from math import pi,sin,cos,sqrt,log,floor
```

```python
cmap=cm.RdBu_r
cmap.set_under('w')

cmapseismic=cm.seismic
cmapseismic.set_under('w')
```

## Visualizing Planck Map

```python
planck_map = hp.read_map('COM_CMB_IQU-smica_2048_R3.00_full.fits')
fig=plt.figure(figsize=(10,6))
hp.mollview(planck_map,min=-0.0003,max=0.0003,title='Planck Temperature Map',fig=1,unit='Temperature (K)',cmap=cmap)
```

```python
plt.xlim(-0.0005,0.0005)

# best fit of data
(mu,sigma) = norm.fit(planck_map)

# the histogram of the dataa
n,bins,patches=plt.hist(planck_map,1500,density=True,alpha=0.25)

# add a 'best fit' line
y = norm.pdf(bins,mu,sigma)
l = plt.plot(bins,y,'k--',lw=2)

plt.xlabel('Temperature (K)')
plt.ylabel('Frequency')
plt.title(r'Histogram of the Planck map')

plt.show()
```

```python
white_noise = np.ma.asarray(np.random.normal(0,sigma,12*2048**2))
fig=plt.figure(figsize=(10,6))
hp.mollview(white_noise,min=-0.0003,max=0.0003,title='White Noise Map',fig=1,unit='Temperature (K)',cmap=cmap)
```

```python
plt.xlim(-0.0005,0.0005)

# best fit of data
(mu,sigma) = norm.fit(white_noise)

# the histogram of the dataa
n,bins,patches=plt.hist(white_noise,100,density=True,alpha=0.25)

# add a 'best fit' line
y = norm.pdf(bins,mu,sigma)
l = plt.plot(bins,y,'k--',lw=2)

plt.xlabel('Temperature (K)')
plt.ylabel('Frequency')
plt.title(r'Histogram of the White Noise')

plt.show()
```
